```python
# Tara Ghose with collaborator Keerthana
# Import necessary packages
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

# Prompt user for URL input
photoInput = input("Please enter the URL of a jpeg photo! ")
photo = io.imread(photoInput)
type(photo)

# Create red_filtered
def red_filtered(photo):
  # Make a copy of orginal photo
    red_filtered_photo = np.copy(photo)
    # Acessing the third dimension at index 1 will set all G to zero
    red_filtered_photo[:, :, 1] = 0
    # Acessing the third dimension at index 2 will set all B to zero
    red_filtered_photo[:, :, 2] = 0
    #print final result (all R)
    return red_filtered_photo

# Create green_filtered
def green_filtered(photo):
  # Make a copy of orginal photo
    green_filtered_photo = np.copy(photo)
    # Acessing the third dimension at index 0 will set all R to zero
    green_filtered_photo[:, :, 0] = 0
    # Acessing the third dimension at index 2 will set all B to zero
    green_filtered_photo[:, :, 2] = 0
    #print final result (all R)
    return green_filtered_photo

# Create red_filtered
def blue_filtered(photo):
  # Make a copy of orginal photo
    blue_filtered_photo = np.copy(photo)
    # Acessing the third dimension at index 1 will set all R to zero
    blue_filtered_photo[:, :, 0] = 0
    # Acessing the third dimension at index 1 will set all G to zero
    blue_filtered_photo[:, :, 1] = 0
    #print final result (all R)
    return blue_filtered_photo

# Return original photo
print("Original photo: ")
plt.imshow(photo)
plt.show()

# Version #1: Call red_filtered function on photoInput
print("Red filtered photo: ")
plt.imshow(red_filtered(photo))
plt.show()

# Version #2: Call green_filtered function on photoInput
print("Green filtered photo: ")
plt.imshow(green_filtered(photo))
plt.show()

# Version #3: Call blue_filtered function on photoInput
print("Blue filtered photo: ")
plt.imshow(blue_filtered(photo))
plt.show()

# Version #4: Call red_filtered and green_filtered function on photoInput
print("Red + green filtered photo: ")
plt.imshow(red_filtered(photo) + green_filtered(photo))
plt.show()

# Version #5: Call red_filtered and blue_filtered function on photoInput
print("Red + blue filtered photo: ")
plt.imshow(red_filtered(photo) + blue_filtered(photo))
plt.show()

# Version #6: Call green_filtered and blue_filtered function on photoInput
print("Green + blue filtered photo: ")
plt.imshow(blue_filtered(photo) + green_filtered(photo))
plt.show()

# Obtain the RBG values of the photo
red_value = red_filtered(photo)
green_value = green_filtered(photo)
blue_value = blue_filtered(photo)
new_values = []

# Add new values to array
for a, b, c, in zip (red_value, green_value, blue_value):
  new_values.append(a+b+c)

# Make array with original values
original_values = np.copy(photo)
arrays_equal = np.array_equal(new_values, original_values)

# Check if arrays are equal
if arrays_equal:
  print("The images are the same! ")
else:
  print("The images are not the same! ")


```
